```python
def filter_shp(shape, min_lon, max_lon, min_lat, max_lat):
    cpoints = np.clip(shape.points, [min_lon, min_lat], [max_lon, max_lat])

    outlines = []
    s = shape.parts[0]
    for p in shape.parts[1:]:
        outline = simplify(cpoints[s:p-1])
        if outline:
            outlines.append(outline)
        s = p
    outline = simplify(cpoints[s:])
    if outline:
        outlines.append(outline)

    return outlines

def simplify(points, md2=2.0):
    out = []
    px, py = points[0]
    for x, y in points[1:]:
        if (x-px)**2 + (y-py)**2 > md2:
            out.append((px, py))
            px, py = x, y
    if __builtin__.all(x == px for x,y in out) or __builtin__.all(y == py for x,y in out):
        # all points fall on boundary?
        return None
    if len(out) <= 2:
        return None
    out.append((px, py))
    return out
```

```python
map1 = filter_shp(shape, -180, -50, -30, 70)
map2 = filter_shp(shape, -25, 75, -30, 70)
map3 = filter_shp(shape, 75, 180, -20, 30)
```
